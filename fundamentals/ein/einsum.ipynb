```python
import torch
import einops as e
```

```python
A = torch.tensor([
    [1, 3, 1],
    [2, 2, 2],
    [5, 5, 5],
])

B = torch.tensor([
    [0, 1, 7],
    [1, 1, 0],
    [1, 1, 1],
])
```

```python
e.einsum(A, B, 'i j, k j -> i k')
```

```python
A = torch.arange(10)
B = torch.arange(5, 15)
print(f'A: {A}')
print(f'B: {B}')
print('-' * 40)

print(e.einsum(A, 'i ->'))
print(e.reduce(A, 'i ->', 'sum'))
print('-' * 40)

print(e.einsum(A, B, 'i, i -> i'))
print('-' * 40)

print(e.einsum(A, B, 'i, i -> '))
print('-' * 40)

print(e.einsum(A, B, 'i, j -> i j'))
print('-' * 40)

C = torch.arange(25).reshape((5, 5))
print(f'C: {C}')
print(e.einsum(C, 'i i -> '))
```

```python
D = e.repeat(torch.arange(-2, 3), 'i -> 5 i')
print(f'D: {D}')
e.einsum(C, D, 'i j, j i -> i j')
```

```python
e.einsum(C, D, 'i j, k l -> i j k l')
```
